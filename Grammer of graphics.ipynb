# Graphics
---

### Data Types
- **Numeric** (AKA quantitative)
  + Continuous
  + Count (AKA Discrete)
- **Categorical** (AKA Qualitative)
  + Binary 
  + Non Binary (Nominal and Ordinal)

### Univariate analysis
- Numeric
  + Distribution (**Histogram**)
    + Normal
      + Minimum, Maximum, Mean, Standard deviation (represents central 68% of data, 2SD=95% of central data, 3SD=99.7% of central data)
    + Non normal 
      + 5 number summary (**Boxplot**) -> Miniumum, Maximum, Median, First Quartile(Q1) and Third Quartile (Q3), Range, Inter quartile range (central 50% of data)
- Categorical (**Bar chart**) 
  + Levels
  + Missingness
  + Frequency
  + Proportions
  + Percentages

### Bivariate analysis
- Categorical vs Categorical
  + 2 way table (**Grouped bar charts or Stacked bar charts**)
- Categorical vs Numeric 
  + Numeric summaries at each categorical level (**Grouped box plots**)
- Numeric vs Numeric
  + Correlation (**Scatter plot**) 






```r
install.packages("Hmisc")
```

```r
library(tidyverse)
library(Hmisc)
```

```r
list(
  id=1:10000, 
  age = rexp(10000, 0.2), 
  sex = sample(c('male','female'), 10000, replace=T, prob=c(0.46, 1-0.46)), 
  weight = rnorm(10000, 70, 10), 
  education = sample(c('elementary', 'high school', 'assoc degree', 'college', 'graduate'), 10000, replace=T, prob=c(0.11, 0.25, 0.22, 0.41, 1-(0.11+0.25 +0.22 +0.41)))
) %>% 
  as.data.frame() -> explore
```

```r
age=rnorm(30, 5, 10000)
```

```r
explore %>% mutate(age = round(age)) %>% 
  mutate(age = (age - min(age))/(max(age)-min(age))) %>% 
  mutate(age = round((age*100)+18)) %>% 
  mutate(weight = round(weight, 2)) -> explore
```
