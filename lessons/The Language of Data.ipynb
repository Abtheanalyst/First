### Speaking to data
- Things we want to do may include:
1. Know how many columns there are
2. Know how many rows there are
3. Select only certain columns eg we may want to select only the serial number and age
4. Filter rows that meet a certain criterion or criteria example we may want to filter for only females

```R
df %>% 
  select(SN, Age) %>% 
  filter(Sex == 'Female')
```


5. Sort your data that is order by a certain variable
6. create new variables from the ones that exist
7. Do some aggregate queying of your data - grouping by certain values on certain columns 

```R
df %>% 
  arrange(Age) %>% 
  mutate(age_in_5_yrs = Age + 5) %>% 
  group_by(Sex) %>% 
  summarize(Counts = n())
```

## Structured Query Language (SQL)
This is the language of data!
in basic form 

```SQL
SELECT column 
FROM table
```

```R
table %>% 
  select(column)
```


```SQL
SELECT column 
FROM table 
WHERE column = 'value'
```

```R
table %>% 
  select(column) %>% 
  filter(column == 'value')
```

```SQL 
SELECT column1, column2 
FROM table 
ORDER BY column2
```

```R
table %>% 
  select(column1, column2) %>% 
  arrange(column2)
```
