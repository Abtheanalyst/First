# Data analytics 

### First steps 
- Know the data including source 
- Know what documentation exists for the data 
  + Code books 
  + Data dictionaries 
- Data types (variables, or columns)
  + This includes the number of variables, their names and their types
- How many records are there (observations, or rows)
### Data Types
- Qualitative (Categorical, Characters, Words)
  + **Binary** (2 levels only) - example - dead or alive, male or female, inpatient or outpatient, etc
  + **Non Binary** - Has more than 2 levels
    + **Nominal** - Named data with no inherent order except alphabetical eg - Name, location, address, marital status etc 
    + **Ordinal** - Categorical data which is non binary and has inherent order eg - low, medum, high; poor, average, good, excellent; etc
- Quantitative (Numerical)
  + **Continuous** - there is no clear sepaeration between any 2 levels or there are too many levels eg weight, height, medical test results etc
  + **Count**(Discrete) where there are clear distinctions between levels or not too many levels example number of products bought, days on admission in hospital, number of times taken overtime, etc


### Univariate analysis
When a new data set is obtained analyse every variable one at a time. This begins with knowing how many values are missing.

- Analysing Single Numeric data 
  + Five (5) number summary (visualized as a boxplot)
    + Minimum
    + First Quartile (Q1)
    + Median
    + Third Quartile (Q3)
    + Maximum value
  + Distribution - visual (histograms) - to determine normality
  + Center and Spread
    + Not normal in distribution - **center** = median, **spread** = inter quartile range - represents the central 50% of the data
    + Normally distributed data - **center** = mean (coincides with the median), **spread** = standard deviation - 1 standard deviation represents 68% of the central data
- Analysing Single Categorical data 
  + What values exisit for this variable
  + How many are they
  + What is the count of each value in the variable -> Frequency
  + What is the proportion of each value in the varibale -> proportions/percentages

**Mean Formula**
$$
\bar{x} = {1\over{n}}{\sum_{i=1}^{n}(x_i)}
$$

**Standard Deviation Formula**
$$
\sigma = \sqrt{{\sum_{i=1}^{n}(\bar{x} - x_i)^2}\over{n-1}}
$$

**68-95-99.7 Rule**
- 1 standard deviation represents the central 68% of a normally distributed data variable
- 2 standard deviations represents the central 95% 
- 3 standard deviations represents the central 99.7%

# Probabilities
- Rule1: $0 \le P(X) \le 1$
- Rule2: $\sum_{i=0}^{n}{P(X_i)} = 1$
- Rule3: $P(A\cup{B}) = P(A) + P(B) - P(A\cap{B})$

### Marginal Probabilities
- Refer to probabilities of a single variable

### Joint Probabilities
- Refer to probabilities accross variables
- Written as $P(A\cap{B})$

### Conditional Probabilities
- Probability of a variable calculated on one level of another variable
- Written as $P(A|B)$
- Read as probability of **A given B**

### Probability Rules continued
- Rule4: P(A|B) = $P(A\cap{B})\over{P(B)}$ 
- Rule4: $P(A\cap{B})$ = $P(A|B) * P(B)$
- Rule4(independence): $P(A\cap{B})$ = $P(A) * P(B)$

### Bivariate analysis

- We are interested in one variable per time: the dependent variable. We are often interested in the summary of the dependent variable as well as how other variables influence it. The other variables are usually referred to as the independent variables

- Scenarios (putting dependent variable first)
  + Binary vs Binary 
  + Binary vs Categorical
  + Categorical vs Categorical
  + Binary vs Continuous
  + Categorical vs Continuous
  + Continuous vs Continuous

##### The 2-way table (2 x 2 table)
- This enables reporting of frequencies and proportions/percentages hence probailities
- Visualize using grouped or stacked bar charts
- Correlation

##### Correlation
This ranges from -1 through 0 to +1 where -1 indicates a strong negative correlation meaning that the variables trend in opposite directions while a positive correlation means that the variables trend in the same direction
